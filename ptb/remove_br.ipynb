```python
import pandas as pd
```

```python
# read the data from pivot tables to data frames

df_o = pd.read_csv('pivot_o.csv')
df_f = pd.read_csv('pivot_f.csv')
df_g = pd.read_csv('pivot_g.csv')
```

```python
df_g
```

```python
# renaming columns to remove '[' and ']' symbols from column names

df_o.rename(columns=lambda x: x.replace('[','').replace(']',''), inplace=True)
df_f.rename(columns=lambda x: x.replace('[','').replace(']',''), inplace=True)
df_g.rename(columns=lambda x: x.replace('[','').replace(']',''), inplace=True)
```

```python
# creating new dataframes with merged identical column names (created after removing the brackets)

df_o_norm = df_o.groupby(level=0, axis=1).sum()
df_f_norm = df_f.groupby(level=0, axis=1).sum()
df_g_norm = df_g.groupby(level=0, axis=1).sum()
```

```python
df_g_norm.iloc[:,1:]
```

```python
# normilizing the data across each row

df_o_norm.iloc[:,1:] = df_o_norm.iloc[:,1:].div(df_o_norm.sum(axis=1), axis=0)
df_f_norm.iloc[:,1:] = df_f_norm.iloc[:,1:].div(df_f_norm.sum(axis=1), axis=0)
df_g_norm.iloc[:,1:] = df_g_norm.iloc[:,1:].div(df_g_norm.sum(axis=1), axis=0)
```

```python
df_o_norm
```

```python
# sorting the column names in the alphabet order

df_o_norm.sort_index(axis=1, inplace=True)
df_f_norm.sort_index(axis=1, inplace=True)
df_g_norm.sort_index(axis=1, inplace=True)
```

```python
# storage normalized data with removed brackets to new csv files

df_o_norm.to_csv('pivot_o_normalized.csv', sep=';', index=False)
df_f_norm.to_csv('pivot_f_normalized.csv', sep=';', index=False)
df_g_norm.to_csv('pivot_g_normalized.csv', sep=';', index=False)
```
